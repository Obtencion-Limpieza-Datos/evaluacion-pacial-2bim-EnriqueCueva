```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("datos/datoscanton.csv", low_memory=False, encoding='utf-8', sep="|")
```

```python
data.keys()
```

```python
data
```

```python
ax = data.groupby([' Partido ']).size().sort_values(ascending=False)[:20].plot.barh(title = u'Partidos con mas candidatos', legend=True)

for p in ax.patches:
    # ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
    ax.annotate(str(p.get_height()), (p.get_x(), p.get_height()))
```
